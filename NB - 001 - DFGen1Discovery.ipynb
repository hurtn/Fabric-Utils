```python
import sempy_labs.admin as labs_admin
import sempy_labs as labs
import pandas as pd
import sempy.fabric as fabric

#def get_token(audience="pbi"):
#    return notebookutils.credentials.getToken(audience)

#client = fabric.FabricRestClient(token_provider=get_token)
client = fabric.FabricRestClient()
data=[]
for i,p in labs_admin.list_workspaces().iterrows():
    if p['Type'] == 'Workspace':
        #print(p['Name'] + p['Type'])
        try:
            response = client.get(f"/v1.0/myorg/groups/{p['Id']}/dataflows").json()
            if len(response['value'])>0:
                if response['value'][0]['generation']==1:
                    if plogging:
                        print(f"Workspace {p['Name']} has gen 1 dataflow named {response['value'][0]['name']}")
                    data.append({
                        "Workspace Name": p['Name'],
                        "Dataflow Name": response['value'][0]['name'],
                        "Generation": response['value'][0]['generation']})
        except Exception as e:
            None
            #print(f"Could not check workspace {p['Name']} due to error:{e}")
df = pd.DataFrame(data)

if not df.empty:
    display(data)
else:
    print('No gen1 dataflows found.')
```
